### **1. Graphs**

A graph $G$ consists of two finite sets: a nonempty set of $V(G)$ vertices and a set $E(G)$ of edges, where each edge is associated with a set consisting of either one or two vertices called its endpoints. The correspondence from edges to endpoints is called the edge-endpoint function.

```python
G = nx.Graph()
G.add_edges_from([(1, 1), (1, 2), (2, 3), (2, 3), (3, 4), (4, 5), (5, 6), (6, 1)])

nx.draw(G, **draw_params)
plt.show()
```

### **2. Directed Graph**

---

A directed graph, or digraph, consists of two finite sets: a nonempty set $V(G)$ of vertices and a set $D(G)$ of directed edges, where each is associated with an ordered pair of vertices called its endpoints. If edge $e$ is associated with the pair $(v,w)$ of vertices, then $e$ is said to be the (directed) edge from $v$ to $w$.

```python
G_directed = nx.DiGraph()
G_directed.add_edges_from([(1, 2), (2, 3), (3, 5), (3, 1), (4, 5), (5, 1)])

nx.draw(G_directed, **draw_params, arrowsize=40)
plt.show()
```

### **3. Simple Graph**

---

A simple graph is a specific type of graph that adheres to certain rules, distinguishing it from other types of graphs. Here are some key differences between a simple graph and a general graph:

- **No self-loops**: In a simple graph, there are no edges that connect a vertex to itself. In contrast, a general graph may allow self-loops, where a vertex is connected to itself by an edge.

- **No multiple edges**: In a simple graph, there is at most one edge between any pair of vertices. This means that if two vertices are connected, they are connected by a single edge. General graphs, however, may allow multiple edges between the same pair of vertices.

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

nx.draw(G, **draw_params)
plt.show()
```

### **4. Complete Graph**

---

A complete graph is a type of graph where each pair of distinct vertices is connected by a unique edge. In other words, in a complete graph, there is an edge between every pair of vertices in the graph.

```python
G_complete = nx.complete_graph(7)

nx.draw(G_complete, **draw_params)
plt.show()
```

### **5. Bipartite Graph**

---

A bipartite graph is a type of graph in which the set of vertices can be divided into two disjoint sets such that no two vertices within the same set are adjacent. In other words, a bipartite graph is a graph whose vertices can be partitioned into two sets such that all edges connect a vertex from one set to a vertex in the other set.

```python
G_bipartite = nx.Graph()
G_bipartite.add_nodes_from([1, 2, 3, 4], bipartite=0)
G_bipartite.add_nodes_from(["A", "B", "C"], bipartite=1)
G_bipartite.add_edges_from(
    [(1, "A"), (1, "B"), (2, "A"), (2, "B"), (2, "C"), (3, "A"), (4, "C")]
)

nx.draw(G_bipartite, nx.bipartite_layout(G_bipartite, [1, 2, 3, 4]), **draw_params)
plt.show()
```

### **6. Subgraphs**

---

A subgraph is a graph that is derived from a larger graph by selecting a subset of its vertices and edges. In other words, a subgraph is a graph that contains a subset of the vertices and edges of the original graph.

```python
_, axes = plt.subplots(1, 2, figsize=(12, 6))

G_main = nx.complete_graph(5)
G_sub = nx.subgraph(G_main, [0, 1, 2])

nx.draw(G_main, ax=axes[0], **draw_params)
axes[0].set_title("Original Graph")

nx.draw(G_sub, ax=axes[1], **draw_params)
axes[1].set_title("Subgraph")

plt.tight_layout()
plt.show()
```

### **8. The Handshake Theorem**

---

The handshake theorem, also known as the handshaking lemma, is a fundamental result in graph theory that relates the sum of the degrees of all vertices in a graph to the total number of edges in the graph. The theorem is named after the concept of a "handshake" between vertices, analogous to people shaking hands in a social gathering.

Here's the formal statement of the handshake theorem:

In any finite undirected graph, the sum of the degrees of all vertices is equal to twice the number of edges.

Mathematically, if $G=(V,E)$ is an undirected graph with $n$ vertices and $m$ edges, where $deg(v)$ denotes the degree of vertex $v$ (the number of edges incident to $v$), then the handshake theorem states:

$
\begin{equation}
	\Sigma_{v \isin V} deg(v)=2m
\end{equation}
$

In other words, the total sum of the degrees of all vertices in the graph is equal to twice the number of edges.

### **9. Travelling in a Graph**

---


In graph theory, there are several ways to travel or traverse a graph, each with its own characteristics and applications.

- **Walk**: A walk is a sequence of vertices and edges, where vertices and edges can be repeated.
- **Trail**: A trail is a walk in which no edge is repeated, though vertices can be repeated.
- **Path**: A path is a sequence of edges that connects a sequence of distinct vertices.
- **Closed Walk**: A closed walk is a walk from $v$ to $w$ that starts and ends in the same vertex.
- **Circuit**: A circuit is a trail that starts and ends at the same vertex.
- **Cycle**: A cycle is a path (no vertices repeated) that starts and ends at the same vertex.

### **10. Connectedness of a Graph**

---

The connectedness of a graph refers to the property of the graph in which every pair of vertices in the graph is connected by a path. In a connected graph, there is at least one path between every pair of vertices, meaning that you can travel from any vertex to any other vertex by following the edges of the graph.

Formally, a graph is considered connected if there exists a path between every pair of vertices in the graph. This path does not need to be direct; it can consist of multiple edges connecting intermediate vertices.

### **11. Euler's Circuit**

---

Euler's circuit, named after the Swiss mathematician Leonhard Euler, is a type of circuit in a graph theory. Specifically, an Euler's circuit is a circuit that traverses every edge of a graph exactly once and returns to the starting vertex, forming a closed loop.

Formally, an Euler's circuit in a graph is a circuit that starts and ends at the same vertex and includes every edge of the graph exactly once.

Key properties of Euler's circuits include:

- **Every edge visited exactly once**: In an Euler's circuit, each edge of the graph is traversed exactly once. This means that the circuit includes every edge in the graph.
- **Returns to the starting vertex**: Euler's circuit starts and ends at the same vertex, forming a closed loop. This means that the circuit returns to the starting vertex after traversing all edges.

```python
G_euler_circuit = nx.cycle_graph(7)
nx.draw(G_euler_circuit, nx.circular_layout(G_euler_circuit), **draw_params)
plt.show()
```

### **12. Euler's Trail**

---


An Euler's trail, also known as an Eulerian trail, is a type of path in a graph theory. Unlike an Euler's circuit, which forms a closed loop, an Euler's trail is a path that traverses every edge of a graph exactly once but may or may not return to the starting vertex.

Formally, an Euler's trail in a graph is a path that includes every edge of the graph exactly once.

Key properties of Euler's trails include:

- **Every edge visited exactly once**: In an Euler's trail, each edge of the graph is traversed exactly once. This means that the trail includes every edge in the graph.
- **May not return to the starting vertex**: Unlike Euler's circuits, which start and end at the same vertex, Euler's trails may start and end at different vertices. However, if an Euler's trail does return to the starting vertex, it is called an Euler's circuit.

```python
G_euler_trail = nx.path_graph(7)
nx.draw(G_euler_trail, nx.circular_layout(G_euler_trail), **draw_params)
plt.show()
```

### **13. Hamiltonian Circuit**

---

A Hamiltonian circuit, named after the Irish mathematician William Rowan Hamilton, is a type of circuit in graph theory. A circuit is a closed path in a graph that visits every vertex exactly once and returns to the starting point. Specifically, a Hamiltonian circuit is a circuit that traverses every vertex of a graph exactly once and returns to the starting vertex, forming a closed loop.

Formally, a Hamiltonian circuit in a graph is a circuit that starts and ends at the same vertex and includes every vertex of the graph exactly once.

Key properties of Hamiltonian circuits include:

- **Every vertex visited exactly once**: In a Hamiltonian circuit, each vertex of the graph is visited exactly once. This means that the circuit includes every vertex in the graph.
- **Returns to the starting vertex**: Hamiltonian circuits start and end at the same vertex, forming a closed loop. This means that the circuit returns to the starting vertex after visiting all other vertices.


```python
G_hamiltonian = nx.cycle_graph(8)
nx.draw(G_hamiltonian, nx.circular_layout(G_hamiltonian), **draw_params)
plt.show()
```
